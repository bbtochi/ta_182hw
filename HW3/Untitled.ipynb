```python
%matplotlib inline
import csv
import pandas as pd
import matplotlib.pyplot as plt
pd.set_option('display.width', 500)
pd.set_option('display.max_columns', 100)
pd.set_option('display.notebook_repr_html', True)
```

```python
sd_df = pd.read_csv('sdf.csv')
trials = []
for i in range(1,6):
    trials.append(sd_df[sd_df['trial'] == i])
```

```python
c = ['r','b','g','y','r']
l = ['red','blue','green','yellow','black']
plt.hold()
for i in range(5):
    df1 = trials[i]
    x1,y1 = df1.iteration.values, df1.conflicts.values
    plt.plot(x1,y1, color=l[i], label='trial '+str(i+1))

plt.legend()
plt.title('regular descent trials')
plt.ylabel('conflicts')
plt.xlabel('interations')
```
